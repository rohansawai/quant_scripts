```python
import numpy as np
```

```python
def ext_rev(df: pd.DataFrame, lookback = 21):
    #candlestick body
    df = df.assign(c_body=df["close"] - df["open"], c_range=df["high"]-df["low"])
    
    #avg body size
    df["avg_candle_size"] = df["c_body"].abs().rolling(window=lookback).mean()
    
    #body %
    df["body_pct"] = df["c_body"].abs() / df["c_range"]
    
    #candle color
    df["c_color"] = df["c_body"].apply(np.sign)
    
    #candle body > 2 * avg candle size
    df["c_1"] = df["c_body"].abs()>df["avg_candle_size"]*2
    
    #body:range bw 0.5 to 0.8
    df["c_2"] = (df["body_pct"] > 0.5) & (df["body_pct"] < 0.85)
    
    #today's candle and yesterday's candle opposite
    df["c_3"] = df["c_color"] != df["c_color"].shift()
    
    #combine all conditions
    df["ext_rev"] = df["c_1"].shift() & df["c_2"].shift() & df["c_3"]
    
    df["ext_rev"] = df["ext_rev"]*df["c_color"]
    return df["ext_rev"]
```

```python
ohlc["ext_rev"] = ext_rev(ohlc)
```

```python
import matplotlib.pyplot as plt
import mplfinance as mpf
import pandas as pd
import matplotlib.dates as mpdates
```

```python
bullish_reversal = ohlc["ext_rev"].replace([-1,0], np.nan)
bearish_reversal = ohlc["ext_rev"].replace([1,0], np.nan)

addplot = [
    mpf.make_addplot(
    bullish_reversal*ohlc["low"]*0.99,
    type="scatter",
    marker="^",
    markersize=250
    ),
    mpf.make_addplot(
    -bearish_reversal*ohlc["high"]*1.01,
    type="scatter",
    marker="v",
    markersize=250
    )    
]
mpf.plot(ohlc, addplot=addplot, style="yahoo", type="candle")
```

```python
from datetime import time

entry_time = time(10, 15)

signal = (
    ohlc[ohlc.index.time < entry_time]["ext_rev"]
    .resample("D")
    .last()
    .dropna()
)

entry_price = (
    ohlc[ohlc.index.time >= entry_time]["open"].resample("D").first().dropna()
)

exit_price = (
    ohlc[ohlc.index.time >= entry_time]["close"].resample("D").last().dropna()
)
```

```python
tradelog = pd.concat(
    [signal, entry_price, exit_price],
    keys=["signal", "entry_price", "exit_price"],
    axis=1
)

tradelog = tradelog[tradelog["signal"]!=0].dropna()

tradelog["pnl"] = (tradelog["exit_price"]-tradelog["entry_price"])*tradelog["signal"]

tradelog["returns"] = tradelog["pnl"]/tradelog["entry_price"]

tradelog["returns"].cumsum().plot()
```

```python
def calculate_metrics(returns:pd.Series):
    returns - returns.dropna()
    
    m = pd.DataFrame(columns=["value"])
    m.loc["gross_return"] = returns.sum()
    
    m.loc["total_trades"] = len(returns)
    m.loc["winning_trades"] = (returns>0).sum()
    m.loc["losing_trades"] = (returns<=0).sum()
    m.loc["avg_profit_winner"] = returns[returns>0].mean()
    m.loc["avg_loss_loser"] = returns[returns<=0].mean()
    m.loc["avg_profit"] = returns.mean()
    
    m.loc["win_rate"] = sum(returns>0)/len(returns)
    m.loc["win_loss_ratio"] = (
        m.loc["avg_profit_winner"]/m.loc["avg_loss_loser"].abs()
    )
    return m

calculate_metrics(tradelog["returns"])
```
